```python


# ============================================================
# 1. Copias de seguridad
# ============================================================

df_original = df.copy()
df_aumentado = df_2.copy()


# ============================================================
# 2. Limpieza básica de nombres de columnas
# ============================================================

df_original.columns = df_original.columns.str.strip()
df_aumentado.columns = df_aumentado.columns.str.strip()

# Eliminar posibles columnas índice tipo Unnamed: 0
df_original = df_original.loc[:, ~df_original.columns.str.contains("^Unnamed")]
df_aumentado = df_aumentado.loc[:, ~df_aumentado.columns.str.contains("^Unnamed")]


# ============================================================
# 3. Comprobar columnas
# ============================================================

if set(df_original.columns) != set(df_aumentado.columns):
    print("Columnas en df_original pero no en df_aumentado:")
    print(set(df_original.columns) - set(df_aumentado.columns))

    print("\nColumnas en df_aumentado pero no en df_original:")
    print(set(df_aumentado.columns) - set(df_original.columns))

    raise ValueError("Las columnas no coinciden.")

# Reordenar df_aumentado con el mismo orden de columnas que df_original
df_aumentado = df_aumentado[df_original.columns]


# ============================================================
# 4. Asegurar video_id como string
# ============================================================

df_original["video_id"] = df_original["video_id"].astype(str)
df_aumentado["video_id"] = df_aumentado["video_id"].astype(str)


# ============================================================
# 5. Quedarse SOLO con las filas sintéticas del archivo aumentado
# ============================================================

df_sinteticos = df_aumentado[
    df_aumentado["video_id"].str.startswith("SYN_")
].copy()


# ============================================================
# 6. Concatenar original + sintéticos
# ============================================================

df_final = pd.concat(
    [df_original, df_sinteticos],
    ignore_index=True
)


# ============================================================
# 7. Eliminar duplicados por seguridad
# ============================================================

df_final = df_final.drop_duplicates(subset=["video_id"], keep="first")


# ============================================================
# 8. Comprobaciones finales
# ============================================================

print("Shape df original:", df_original.shape)
print("Shape df aumentado:", df_aumentado.shape)
print("Shape sintéticos añadidos:", df_sinteticos.shape)
print("Shape df final:", df_final.shape)

print("\nVídeos sintéticos en df_final:")
print(df_final["video_id"].str.startswith("SYN_").sum())

print("\nVídeos reales en df_final:")
print((~df_final["video_id"].str.startswith("SYN_")).sum())

print("\n¿Columnas finales iguales a las originales?")
print(df_final.columns.tolist() == df_original.columns.tolist())

display(df_final.head())
display(df_final[df_final["video_id"].str.startswith("SYN_")].head())

```

```python
def tipo_variables(df):
    dtypes = df.dtypes
    categorical_cols = []
    numerical_cols = []

    for col, dtype in dtypes.items():
        if pd.api.types.is_numeric_dtype(dtype):
            numerical_cols.append(col)
        else:
            categorical_cols.append(col)

    return numerical_cols, categorical_cols

numerical_cols, categorical_cols = tipo_variables(df)
```

# 8 Variable objetivo

```python
# ============================================================
# Estimación de visualizaciones a 7 días
# ============================================================

# Copia de trabajo
df = df.copy()

# ------------------------------------------------------------
# 1. Filtrar vídeos con antigüedad entre 6 y 30 días
# ------------------------------------------------------------

df = df[df["video_age_days"].between(6, 30)].copy()

# ------------------------------------------------------------
# 2. Crear edad redondeada del vídeo
# ------------------------------------------------------------

df["age_day"] = df["video_age_days"].round().astype(int)
df["age_day"] = df["age_day"].clip(lower=6, upper=30)

# ------------------------------------------------------------
# 3. Crear logaritmo de las visualizaciones observadas
# ------------------------------------------------------------

df["log_views_observed"] = np.log1p(df["view_count"])

# ------------------------------------------------------------
# 4. Corrección paramétrica a horizonte de 7 días
# ------------------------------------------------------------

H_DAYS = 7
ALPHA = 0.5

df["age_correction_to_7d"] = ALPHA * np.log(H_DAYS / df["age_day"])

# ------------------------------------------------------------
# 5. Aplicar corrección en escala logarítmica
# ------------------------------------------------------------

df["log_views_7d_est"] = (
    df["log_views_observed"] + df["age_correction_to_7d"]
)

# ------------------------------------------------------------
# 6. Volver a escala original
# ------------------------------------------------------------

df["views_7d_est"] = np.maximum(
    np.expm1(df["log_views_7d_est"]),
    0
)

# ------------------------------------------------------------
# 7. Comprobaciones básicas
# ------------------------------------------------------------

print("Número de vídeos tras filtrar 6-30 días:", len(df))
print("Edad mínima:", df["age_day"].min())
print("Edad máxima:", df["age_day"].max())

df[[
    "age_day",
    "view_count",
    "views_7d_est",
    "age_correction_to_7d"
]].sample(10, random_state=42).sort_values("age_day")
```

```python
# ============================================================
# Definición de variable objetivo con baseline por canal y formato
# ============================================================

df = df.sort_values(["channel_id", "is_short", "published_at_local"]).copy()

H_DAYS = 7
WINDOW_SIZE = 10
MIN_HISTORY = 1

# Usamos la estimación a 7 días como métrica base
df["views_7d"] = df["views_7d_est"]
df["log_views_7d"] = np.log1p(df["views_7d"])
```

```python
def compute_lagged_baseline_7d_by_format(group):
    group = group.sort_values("published_at_local").copy()
    
    baselines = []

    for idx, row in group.iterrows():
        current_time = row["published_at_local"]

        eligible_history = group[
            group["published_at_local"] + pd.Timedelta(days=H_DAYS) <= current_time
        ]

        previous_values = eligible_history["views_7d"].tail(WINDOW_SIZE)

        if len(previous_values) >= MIN_HISTORY:
            baseline = previous_values.median()
        else:
            baseline = np.nan

        baselines.append(baseline)

    group["baseline_views_7d_format"] = baselines
    
    return group
```

```python
# Crear columnas auxiliares solo para agrupar
df["_channel_id_key"] = df["channel_id"]
df["_is_short_key"] = df["is_short"]

# Aplicar baseline por canal + formato usando las columnas auxiliares
df = (
    df.groupby(["_channel_id_key", "_is_short_key"], group_keys=False)
      .apply(compute_lagged_baseline_7d_by_format)
      .reset_index(drop=True)
)

# Eliminar columnas auxiliares
df = df.drop(columns=["_channel_id_key", "_is_short_key"], errors="ignore")
```

```python
df["log_baseline_views_7d_format"] = np.log1p(df["baseline_views_7d_format"])

df["target_relative_log_views_7d"] = (
    df["log_views_7d"] - df["log_baseline_views_7d_format"]
)
```

## Definición de la variable objetivo

Una vez estimadas las visualizaciones a 7 días, se define una variable objetivo relativa al rendimiento histórico del canal. Esta decisión evita modelar directamente las visualizaciones absolutas, que están fuertemente condicionadas por el tamaño del creador.

Para cada vídeo se calcula un baseline histórico como la mediana de visualizaciones estimadas a 7 días de las últimas publicaciones anteriores del mismo canal. Para evitar fuga de información temporal, solo se consideran vídeos cuyo rendimiento a 7 días habría sido observable antes de la publicación actual.

La variable objetivo se define como:

\[
y_i^{(7)} =
\log(1 + \widehat{V}_i^{(7)})
-
\log(1 + B_i^{(7)})
\]

donde \(\widehat{V}_i^{(7)}\) son las visualizaciones estimadas a 7 días y \(B_i^{(7)}\) representa el baseline histórico del canal.

## Estimación de visualizaciones a 7 días

Para aumentar el tamaño muestral se consideran vídeos con una antigüedad entre 6 y 30 días. Dado que las visualizaciones acumuladas dependen del tiempo transcurrido desde la publicación, se transforma el número de visualizaciones observadas a una estimación comparable a un horizonte de 7 días.

Se aplica una corrección paramétrica en escala logarítmica:

\[
\widehat{V}_i^{(7)} = V_i^{(A_i)} \left(\frac{7}{A_i}\right)^\alpha
\]

donde \(A_i\) representa la edad del vídeo en días y \(\alpha\) controla la intensidad de la corrección. En este trabajo se utiliza inicialmente \(\alpha = 0.5\), lo que proporciona una corrección sublineal y conservadora: los vídeos más antiguos se ajustan a la baja, pero no de forma tan agresiva como una división proporcional directa por la edad.

La variable resultante `views_7d_est` se utilizará como aproximación de las visualizaciones a 7 días.

```python
def count_tags(x):
    if pd.isna(x):
        return 0
    
    if isinstance(x, list):
        return len(x)
    
    if isinstance(x, str):
        x = x.strip()
        
        if x == "":
            return 0
        
        # Si los tags vienen separados por coma
        if "," in x:
            return len([tag for tag in x.split(",") if tag.strip() != ""])
        
        # Si vienen como string único
        return len(x.split())
    
    return 0
```

```python
def add_recent_post_counts(df, group_cols, time_col, windows_hours, prefix):
    df = df.copy()
    
    for window in windows_hours:
        df[f"{prefix}_last_{window}h"] = 0

    for _, group in df.groupby(group_cols, sort=False):
        group = group.sort_values(time_col)
        times = group[time_col]

        for idx, current_time in times.items():
            for window in windows_hours:
                start_time = current_time - pd.Timedelta(hours=window)

                count = (
                    (times < current_time) &
                    (times >= start_time)
                ).sum()

                df.loc[idx, f"{prefix}_last_{window}h"] = count

    return df
```

```python
df_fe = add_recent_post_counts(
    df=df_fe,
    group_cols=["channel_id"],
    time_col="published_at_local",
    windows_hours=[24, 48, 168],
    prefix="posts"
)
```

```python
def compute_historical_performance_features(group):
    group = group.sort_values("published_at_local").copy()

    hist_n_valid = []
    hist_median_log_views = []
    hist_mean_log_views = []
    hist_std_log_views = []
    hist_median_target = []
    hist_mean_target = []

    for idx, row in group.iterrows():
        current_time = row["published_at_local"]

        eligible_history = group[
            group["published_at_local"] + pd.Timedelta(days=H_DAYS) <= current_time
        ]

        recent_history = eligible_history.tail(WINDOW_SIZE)

        hist_n_valid.append(len(recent_history))

        if len(recent_history) > 0:
            hist_median_log_views.append(
                recent_history["log_views_7d_est"].median()
            )
            hist_mean_log_views.append(
                recent_history["log_views_7d_est"].mean()
            )
            hist_std_log_views.append(
                recent_history["log_views_7d_est"].std()
            )
        else:
            hist_median_log_views.append(np.nan)
            hist_mean_log_views.append(np.nan)
            hist_std_log_views.append(np.nan)

        valid_targets = recent_history["target_relative_log_views_7d"].dropna()

        if len(valid_targets) > 0:
            hist_median_target.append(valid_targets.median())
            hist_mean_target.append(valid_targets.mean())
        else:
            hist_median_target.append(np.nan)
            hist_mean_target.append(np.nan)

    features = pd.DataFrame(
        {
            "hist_n_valid_format": hist_n_valid,
            "hist_median_log_views_7d_format": hist_median_log_views,
            "hist_mean_log_views_7d_format": hist_mean_log_views,
            "hist_std_log_views_7d_format": hist_std_log_views,
            "hist_median_target_format": hist_median_target,
            "hist_mean_target_format": hist_mean_target,
        },
        index=group.index
    )

    return features
```

```python
hist_feature_frames = []

for _, group in df_fe.groupby(["channel_id", "is_short"], sort=False):
    hist_feature_frames.append(
        compute_historical_performance_features(group)
    )

hist_features_df = pd.concat(hist_feature_frames, axis=0).sort_index()

df_fe = df_fe.join(hist_features_df)
```

# Modelo traint,test,split

```python
model_data = df_fe[all_feature_cols + [target_col, "published_at_local"]].copy()

model_data = model_data.sort_values("published_at_local").reset_index(drop=True)

X = model_data[all_feature_cols]
y = model_data[target_col]
```

```python
n = len(model_data)

train_end = int(n * 0.70)
val_end = int(n * 0.85)

train_data = model_data.iloc[:train_end].copy()
val_data = model_data.iloc[train_end:val_end].copy()
test_data = model_data.iloc[val_end:].copy()

X_train = train_data[all_feature_cols]
y_train = train_data[target_col]

X_val = val_data[all_feature_cols]
y_val = val_data[target_col]

X_test = test_data[all_feature_cols]
y_test = test_data[target_col]

print("Train:", X_train.shape)
print("Validation:", X_val.shape)
print("Test:", X_test.shape)

print("Train fechas:", train_data["published_at_local"].min(), train_data["published_at_local"].max())
print("Val fechas:", val_data["published_at_local"].min(), val_data["published_at_local"].max())
print("Test fechas:", test_data["published_at_local"].min(), test_data["published_at_local"].max())
```

```python
categorical_features = [
    "category_id",
    "channel_country",
    "definition"
]

numeric_features = [
    col for col in all_feature_cols
    if col not in categorical_features
]
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

baseline_pred_val = np.full_like(y_val, y_train.mean(), dtype=float)
baseline_pred_test = np.full_like(y_test, y_train.mean(), dtype=float)

def regression_metrics(y_true, y_pred):
    return {
        "MAE": mean_absolute_error(y_true, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),
        "R2": r2_score(y_true, y_pred)
    }

baseline_val_metrics = regression_metrics(y_val, baseline_pred_val)
baseline_test_metrics = regression_metrics(y_test, baseline_pred_test)

baseline_val_metrics, baseline_test_metrics
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge

ridge_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", Ridge(alpha=1.0))
    ]
)

ridge_model.fit(X_train, y_train)

ridge_val_pred = ridge_model.predict(X_val)
ridge_test_pred = ridge_model.predict(X_test)

ridge_val_metrics = regression_metrics(y_val, ridge_val_pred)
ridge_test_metrics = regression_metrics(y_test, ridge_test_pred)

ridge_val_metrics, ridge_test_metrics
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", RandomForestRegressor(
            n_estimators=300,
            max_depth=8,
            min_samples_leaf=5,
            random_state=42,
            n_jobs=-1
        ))
    ]
)

rf_model.fit(X_train, y_train)

rf_val_pred = rf_model.predict(X_val)
rf_test_pred = rf_model.predict(X_test)

rf_val_metrics = regression_metrics(y_val, rf_val_pred)
rf_test_metrics = regression_metrics(y_test, rf_test_pred)

rf_val_metrics, rf_test_metrics
```

```python
from sklearn.ensemble import ExtraTreesRegressor

extra_trees_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", ExtraTreesRegressor(
            n_estimators=500,
            max_depth=10,
            min_samples_leaf=5,
            random_state=42,
            n_jobs=-1
        ))
    ]
)

extra_trees_model.fit(X_train, y_train)

extra_trees_val_pred = extra_trees_model.predict(X_val)
extra_trees_test_pred = extra_trees_model.predict(X_test)

extra_trees_val_metrics = regression_metrics(y_val, extra_trees_val_pred)
extra_trees_test_metrics = regression_metrics(y_test, extra_trees_test_pred)

extra_trees_val_metrics, extra_trees_test_metrics
```

```python
from xgboost import XGBRegressor

xgb_model = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("model", XGBRegressor(
            n_estimators=500,
            max_depth=4,
            learning_rate=0.03,
            subsample=0.8,
            colsample_bytree=0.8,
            reg_lambda=1.0,
            reg_alpha=0.1,
            objective="reg:squarederror",
            random_state=42,
            n_jobs=-1
        ))
    ]
)

xgb_model.fit(X_train, y_train)

xgb_val_pred = xgb_model.predict(X_val)
xgb_test_pred = xgb_model.predict(X_test)

xgb_val_metrics = regression_metrics(y_val, xgb_val_pred)
xgb_test_metrics = regression_metrics(y_test, xgb_test_pred)

xgb_val_metrics, xgb_test_metrics
```

```python
# ============================================================
# Recomendador top-k de horarios de publicación
# ============================================================

DAY_NAMES_ES = {
    0: "Lunes",
    1: "Martes",
    2: "Miércoles",
    3: "Jueves",
    4: "Viernes",
    5: "Sábado",
    6: "Domingo"
}


def generate_candidate_slots(start_datetime, days_ahead=7, freq="1h"):
    """
    Genera horarios candidatos desde start_datetime durante los próximos days_ahead días.
    """
    start_datetime = pd.Timestamp(start_datetime)

    candidate_times = pd.date_range(
        start=start_datetime,
        periods=days_ahead * 24,
        freq=freq
    )

    return pd.DataFrame({
        "candidate_published_at_local": candidate_times
    })


def compute_history_for_candidate(
    df_history,
    channel_id,
    is_short,
    candidate_time,
    h_days=7,
    window_size=10,
    min_history=1
):
    """
    Calcula features históricas del canal/formato para un horario candidato.
    Solo usa información anterior al candidate_time.
    """

    channel_history = (
        df_history[
            (df_history["channel_id"] == channel_id) &
            (df_history["published_at_local"] < candidate_time)
        ]
        .sort_values("published_at_local")
        .copy()
    )

    format_history = (
        channel_history[channel_history["is_short"] == is_short]
        .sort_values("published_at_local")
        .copy()
    )

    # -------------------------
    # Conteos de publicaciones
    # -------------------------

    channel_posts_before = len(channel_history)
    channel_format_posts_before = len(format_history)

    # -------------------------
    # Tiempo desde último post
    # -------------------------

    if len(channel_history) > 0:
        previous_post_time = channel_history["published_at_local"].max()
        hours_since_last_post = (
            candidate_time - previous_post_time
        ).total_seconds() / 3600
        has_previous_post = 1
    else:
        hours_since_last_post = np.nan
        has_previous_post = 0

    if len(format_history) > 0:
        previous_format_post_time = format_history["published_at_local"].max()
        hours_since_last_format_post = (
            candidate_time - previous_format_post_time
        ).total_seconds() / 3600
        has_previous_format_post = 1
    else:
        hours_since_last_format_post = np.nan
        has_previous_format_post = 0

    # -------------------------
    # Posts recientes
    # -------------------------

    def count_posts_last_hours(history_df, hours):
        start_time = candidate_time - pd.Timedelta(hours=hours)
        return (
            (history_df["published_at_local"] >= start_time) &
            (history_df["published_at_local"] < candidate_time)
        ).sum()

    posts_last_24h = count_posts_last_hours(channel_history, 24)
    posts_last_48h = count_posts_last_hours(channel_history, 48)
    posts_last_168h = count_posts_last_hours(channel_history, 168)

    format_posts_last_24h = count_posts_last_hours(format_history, 24)
    format_posts_last_48h = count_posts_last_hours(format_history, 48)
    format_posts_last_168h = count_posts_last_hours(format_history, 168)

    # -------------------------
    # Histórico válido sin leakage
    # -------------------------

    eligible_history = format_history[
        format_history["published_at_local"] + pd.Timedelta(days=h_days) <= candidate_time
    ].copy()

    recent_history = eligible_history.tail(window_size)

    hist_n_valid_format = len(recent_history)

    if hist_n_valid_format > 0:
        hist_median_log_views_7d_format = recent_history["log_views_7d_est"].median()
        hist_mean_log_views_7d_format = recent_history["log_views_7d_est"].mean()
        hist_std_log_views_7d_format = recent_history["log_views_7d_est"].std()

        hist_median_target_format = recent_history["target_relative_log_views_7d"].dropna().median()
        hist_mean_target_format = recent_history["target_relative_log_views_7d"].dropna().mean()

        baseline_views_7d_candidate = recent_history["views_7d_est"].median()
    else:
        hist_median_log_views_7d_format = np.nan
        hist_mean_log_views_7d_format = np.nan
        hist_std_log_views_7d_format = np.nan
        hist_median_target_format = np.nan
        hist_mean_target_format = np.nan
        baseline_views_7d_candidate = np.nan

    return {
        "channel_posts_before": channel_posts_before,
        "channel_format_posts_before": channel_format_posts_before,

        "hours_since_last_post": hours_since_last_post,
        "has_previous_post": has_previous_post,
        "hours_since_last_format_post": hours_since_last_format_post,
        "has_previous_format_post": has_previous_format_post,

        "posts_last_24h": posts_last_24h,
        "posts_last_48h": posts_last_48h,
        "posts_last_168h": posts_last_168h,

        "format_posts_last_24h": format_posts_last_24h,
        "format_posts_last_48h": format_posts_last_48h,
        "format_posts_last_168h": format_posts_last_168h,

        "hist_n_valid_format": hist_n_valid_format,
        "hist_median_log_views_7d_format": hist_median_log_views_7d_format,
        "hist_mean_log_views_7d_format": hist_mean_log_views_7d_format,
        "hist_std_log_views_7d_format": hist_std_log_views_7d_format,
        "hist_median_target_format": hist_median_target_format,
        "hist_mean_target_format": hist_mean_target_format,

        "baseline_views_7d_candidate": baseline_views_7d_candidate
    }
```

```python
def build_candidate_feature_table(
    df_history,
    base_video_row,
    candidate_times_df,
    all_feature_cols,
    h_days=7,
    window_size=10,
    min_history=1
):
    """
    Crea una tabla de candidatos con las mismas features que espera el modelo.
    El contenido del vídeo se copia desde base_video_row.
    Las variables temporales e históricas se recalculan para cada horario candidato.
    """

    rows = []

    channel_id = base_video_row["channel_id"]
    is_short = base_video_row["is_short"]

    for candidate_time in candidate_times_df["candidate_published_at_local"]:
        row = {}

        # ----------------------------------------------------
        # Copiar por defecto valores del vídeo base
        # ----------------------------------------------------
        for col in all_feature_cols:
            if col in base_video_row.index:
                row[col] = base_video_row[col]
            else:
                row[col] = np.nan

        # ----------------------------------------------------
        # Variables temporales del candidato
        # ----------------------------------------------------
        candidate_time = pd.Timestamp(candidate_time)

        publish_hour = candidate_time.hour
        publish_day_num = candidate_time.dayofweek
        publish_day_name = DAY_NAMES_ES[publish_day_num]

        row["publish_hour_sin"] = np.sin(2 * np.pi * publish_hour / 24)
        row["publish_hour_cos"] = np.cos(2 * np.pi * publish_hour / 24)

        row["publish_day_sin"] = np.sin(2 * np.pi * publish_day_num / 7)
        row["publish_day_cos"] = np.cos(2 * np.pi * publish_day_num / 7)

        row["is_weekend"] = int(publish_day_num in [5, 6])

        # ----------------------------------------------------
        # Variables históricas del candidato
        # ----------------------------------------------------
        history_features = compute_history_for_candidate(
            df_history=df_history,
            channel_id=channel_id,
            is_short=is_short,
            candidate_time=candidate_time,
            h_days=h_days,
            window_size=window_size,
            min_history=min_history
        )

        row.update(history_features)

        # ----------------------------------------------------
        # Columnas auxiliares para interpretar recomendación
        # ----------------------------------------------------
        row["candidate_published_at_local"] = candidate_time
        row["candidate_dayofweek"] = publish_day_name
        row["candidate_hour"] = publish_hour
        row["channel_id"] = channel_id
        row["channel_title"] = base_video_row["channel_title"]
        row["is_short"] = is_short

        rows.append(row)

    candidate_df = pd.DataFrame(rows)

    return candidate_df
```

```python
def impute_candidate_features(candidate_df, reference_df, all_feature_cols, categorical_features):
    """
    Imputa nulos en candidatos usando medianas/modas del dataframe histórico.
    """

    candidate_df = candidate_df.copy()

    for col in all_feature_cols:
        if col not in candidate_df.columns:
            candidate_df[col] = np.nan

    numeric_cols = [
        col for col in all_feature_cols
        if col not in categorical_features
    ]

    for col in numeric_cols:
        if candidate_df[col].isna().any():
            if col in reference_df.columns:
                fill_value = reference_df[col].median()
            else:
                fill_value = 0

            candidate_df[col] = candidate_df[col].fillna(fill_value)

    for col in categorical_features:
        if candidate_df[col].isna().any():
            if col in reference_df.columns:
                mode_values = reference_df[col].dropna().mode()
                fill_value = mode_values.iloc[0] if len(mode_values) > 0 else "Unknown"
            else:
                fill_value = "Unknown"

            candidate_df[col] = candidate_df[col].fillna(fill_value)

        candidate_df[col] = candidate_df[col].astype(str)

    return candidate_df
```

```python
def recommend_top_k_publication_times(
    model,
    df_history,
    base_video_row,
    all_feature_cols,
    categorical_features,
    start_datetime,
    days_ahead=7,
    top_k=3,
    h_days=7,
    window_size=10,
    min_history=1
):
    """
    Devuelve los top-k horarios recomendados para publicar un vídeo.
    """

    candidate_times_df = generate_candidate_slots(
        start_datetime=start_datetime,
        days_ahead=days_ahead,
        freq="1h"
    )

    candidate_df = build_candidate_feature_table(
        df_history=df_history,
        base_video_row=base_video_row,
        candidate_times_df=candidate_times_df,
        all_feature_cols=all_feature_cols,
        h_days=h_days,
        window_size=window_size,
        min_history=min_history
    )

    candidate_df = impute_candidate_features(
        candidate_df=candidate_df,
        reference_df=df_history,
        all_feature_cols=all_feature_cols,
        categorical_features=categorical_features
    )

    # Predicción del target relativo
    candidate_df["predicted_target_relative_log_views_7d"] = model.predict(
        candidate_df[all_feature_cols]
    )

    # Interpretación multiplicativa
    candidate_df["predicted_ratio_vs_baseline"] = np.exp(
        candidate_df["predicted_target_relative_log_views_7d"]
    )

    # Predicción aproximada de views si hay baseline disponible
    candidate_df["predicted_views_7d"] = (
        (1 + candidate_df["baseline_views_7d_candidate"])
        * candidate_df["predicted_ratio_vs_baseline"]
        - 1
    )

    candidate_df = candidate_df.sort_values(
        "predicted_target_relative_log_views_7d",
        ascending=False
    )

    output_cols = [
        "candidate_published_at_local",
        "candidate_dayofweek",
        "candidate_hour",
        "channel_title",
        "is_short",
        "baseline_views_7d_candidate",
        "predicted_target_relative_log_views_7d",
        "predicted_ratio_vs_baseline",
        "predicted_views_7d",
        "hours_since_last_post",
        "hours_since_last_format_post",
        "posts_last_24h",
        "posts_last_48h",
        "posts_last_168h",
        "format_posts_last_24h",
        "format_posts_last_48h",
        "format_posts_last_168h",
        "hist_n_valid_format"
    ]

    return candidate_df[output_cols].head(top_k)
```
